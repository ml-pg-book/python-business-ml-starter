```python
from pathlib import Path
import pandas as pd
```

```python
data_dir = Path("../data")
```

```python
df = pd.read_csv(data_dir / "realestate.csv")
```

```python
df.head()
```

```python
from sklearn.linear_model import Ridge
```

```python
target_col = 'price'
feature_cols = ['square_meter', 'year_from_built', 'distance_from_station']

y_train = df[target_col]
X_train = df[feature_cols]

model = Ridge()
model.fit(X_train, y_train)
```

```python
df["pred"] = model.predict(X_train)
df
```

```python
df.drop(columns=['price']).sample(frac=1).to_csv(data_dir / "realestate_pred.csv", index=False)
```
